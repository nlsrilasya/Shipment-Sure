#### Outlier
It is a data point which is completely different from the reamining observations. It feels like it does not belong to the dataset.

For example: In a dataset of student ages [18, 19, 20, 21, 22, 60], the age 60 is an outlier, because it doesn’t fit the normal range of student ages.
Outliers can skew mean, distort analysis, and mislead models.

#### Skewness 
It tells us about the asymmetry (tilt) of a distribution compared to a normal (symmetrical) distribution.                                                
-> If skewness=0 Data is perfectly symmetrical (like a bell curve).                                                                                      
-> Positive skew(right-skewed) Tail is longer on the right side, meaning more small values and few very large values.                                    
-> Negative skew(left-skewed) Tail is longer on the left side, meaning more large values and few very small values.

#### Quartile: Divides the data into 4 equal parts
Example: If student marks are arranged, Q1 is the point where 25% students scored less, Q2 is the middle, Q3 is where 75% scored less.
#### Percentile: Divides the data into 100 equal parts 
Example: 40th percentile → 40% of data lies below it.                                                                                                    
90th percentile → 90% of data lies below it.

#### Quartiles are just specific percentiles
Q1=25th percentile
Q2=50th percentile (median)
Q3=75th percentile

#### IQR - Inter Quartile Range
It is the range between the first quartile (Q1, 25th percentile) and the third quartile (Q3, 75th percentile).                                           
Formula: IQR=Q3−Q1, Lower Limit=Q1−1.5×IQR, Upper Limit=Q3+1.5×IQR

For Q1 find the median of first half, For Q3 find the median of second half of data.

```python
plt.figure(figsize=(8,5))
df['Weight_in_gms'].plot(kind='box')
plt.title("Outlier Detection using IQR (Weight_in_gms)")
plt.show()
```

-> If the data is positively skewed: Mode<Median<Mean

-> If the data is negatively skewed: Mean<Median<Mode

-> No skewness - Mean=Median=Mode

1. Testing outliers - Identifies data points that deviate significantly from the rest of the dataset.

## 3. Feature scaling and transformation

-> If the features are not scaled there are chances for the model to think that higher values has higher priority(mostly in regression models)           
-> If data is scaled, the gradient descent convergence will be faster.

Scaling can be performed in two ways
1. Standardisation
2. Normalisation

### Standardisation
Standardization is a feature scaling technique where we transform data so that it has: Mean = 0, Standard Deviation = 1

This helps when features are on different scales (e.g., weight in grams vs customer rating out of 5).

Tells how far each value is from mean.

#### Formula: X_new=(X-mean)/std

### Normalisation

Unlike Standardization (which makes mean=0, std=1), Normalization rescales values to a fixed range, usually [0,1]. After normalization, both features will be in the [0,1] range, making them directly comparable without changing their shape.

#### Formula: X'=(X-Xmin)/(Xmax-Xmin)

If there are many outliers then use standardisation. Else you can use anything

### 4. Selecting the right feature 
i. Filter Method
ii. Embedded Method
iii. Wrapper Method

i. Filter method: Features are selected based on statistical measures (like correlation, chi-square test, ANOVA, mutual information) without involving any ML model.

ii. Wrapper method: Uses a predictive model to evaluate different subsets of features by training and testing repeatedly.

iii. Embedded method: Feature selection is performed during model training by the algorithm itself.

### VIF - Variance Inflation Factor
It measures how much a feature (independent variable) is correlated with other features in a dataset.

If VIF = 1 → no correlation with others (good).

If VIF = high (say > 5 or 10) → the feature is strongly related to others (bad, means multicollinearity).

It’s like a warning - “This feature is not giving unique information, it’s repeating what others already said.”

### What Is P-Value?
In statistical hypothesis testing, P-Value or probability value can be defined as the measure of the probability that a real-valued test statistic is at least as extreme as the value actually obtained. P-value shows how likely it is that your set of observations could have occurred under the null hypothesis. P-Values are used in statistical hypothesis testing to determine whether to reject the null hypothesis. The smaller the p-value, the stronger the likelihood that you should reject the null hypothesis

A small p value (<=0.05), you should reject the null hypothesis

A large p value (>0.05), you should not reject the null hypothesis

### Wrapper methods:

1. Forward Selection
Starts with no features.
Adds one feature at a time.
At each step, it adds the feature that improves model performance the most.
Stops when adding more features does not improve the model.
Example use case: When you expect only a few features to be useful and want a quick way to build up a model.

2. Backward Elimination
Starts with all features.
Removes one feature at a time.
At each step, it removes the feature that contributes the least to the model.
Stops when removing more features degrades performance.
Example use case: When you suspect some features are redundant or irrelevant and want to simplify the model.

3. Recursive Feature Elimination (RFE)
A combination of backward elimination with a recursive approach.
Ranks all the features based on their importance.
Recursively removes the least important ones and retrains the model.
Continues until the desired number of features is reached.

### The train-test split method 
It is a fundamental technique in machine learning used to evaluate the performance of a model. It involves dividing a dataset into two distinct subsets: 

Training Set:
This subset is used to train the machine learning model. The model learns patterns and relationships from this data to make predictions or classifications.

Testing Set:
This subset is used to evaluate the performance of the trained model on unseen data. The model's ability to generalize to new, unobserved examples is assessed using this set.
